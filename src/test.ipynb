```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dnaspaces_cleaned.csv")

length = 228.30688
width = 346.92673
height = 13.0
offsetX = 14.67
offsetY = 21.18
imageWidth = 4162
imageHeight = 2739

ids = df['id'].unique()

print(ids)

df
```

```python
df[df['x'] != 0]
```

```python
# find out where the x and y values change
# give indices of the changes
x = df['x'].to_numpy()
y = df['y'].to_numpy()

x_diff = x[1:] - x[:-1]
y_diff = y[1:] - y[:-1]

x_diff = x_diff != 0
y_diff = y_diff != 0

x_diff = x_diff.nonzero()
y_diff = y_diff.nonzero()

print(x_diff)
print(y_diff)
```

```python

# df = df[df['confidence'] < 100]

df = df[df['id'] == ids[0]]
for id in ids:
  data = df[df['id'] == id]
  datax = data['x'].rolling(window=10).mean()
  datay = data['y'].rolling(window=10).mean()
  datax = (datax + offsetX) / (width + 2 * offsetX) * imageWidth
  datay = (datay + offsetY) / (length + 2 * offsetY) * imageHeight
  plt.plot(datax, datay, label=id, marker='o', linestyle='-', markersize=3)

plt.imshow(plt.imread('res/floor1.png'))
plt.savefig('plot.png', dpi=300)
```
